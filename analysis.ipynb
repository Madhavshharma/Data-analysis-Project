<p>Q. how many category are in total orders 


```python
categoryy={}
for category in data["Category"].unique():
    count=0
    for i in data["Category"]:
        if i==category:
            count=count+1
    print(category,count)
    categoryy[category]=count
    
print(categoryy)
```

```python
from sklearn.impute import SimpleImputer
import numpy as np
```

```python
impute=SimpleImputer(missing_values=np.nan,strategy="most_frequent")
impute.fit(data[["currency"]])
```

```python
x=impute.transform(data[["currency"]])
x
data[["currency"]]=x
```
